```python
import numpy as np
import serial
import time
import matplotlib.pyplot as plt
```

```python
ACCData=[0.0]*8
GYROData=[0.0]*8
AngleData=[0.0]*8          
FrameState = 0            #通过0x后面的值判断属于哪一种情况
Bytenum = 0               #读取到这一段的第几位
CheckSum = 0              #求和校验位          
a = [0.0]*3
w = [0.0]*3
Angle = [0.0]*3
```

```python
def DueData(inputdata,Omega,Ang,Acc):   #新增的核心程序，对读取的数据进行划分，各自读到对应的数组里
    global  FrameState    #在局部修改全局变量，要进行global的定义
    global  Bytenum
    global  CheckSum
    global  a
    global  w
    global  Angle
    for data in inputdata:  #在输入的数据进行遍历
        #Python2软件版本这里需要插入 data = ord(data)*****************************************************************************************************
        if FrameState==0:   #当未确定状态的时候，进入以下判断
            if data==0x55 and Bytenum==0: #0x55位于第一位时候，开始读取数据，增大bytenum
                CheckSum=data
                Bytenum=1
                continue
            elif data==0x51 and Bytenum==1:#在byte不为0 且 识别到 0x51 的时候，改变frame
                CheckSum+=data
                FrameState=1
                Bytenum=2
            elif data==0x52 and Bytenum==1: #同理
                CheckSum+=data
                FrameState=2
                Bytenum=2
            elif data==0x53 and Bytenum==1:
                CheckSum+=data
                FrameState=3
                Bytenum=2
        elif FrameState==1: # acc    #已确定数据代表加速度
            
            if Bytenum<10:            # 读取8个数据
                ACCData[Bytenum-2]=data # 从0开始
                CheckSum+=data
                Bytenum+=1
            else:
                if data == (CheckSum&0xff):  #假如校验位正确
                    a = get_acc(ACCData)
                CheckSum=0                  #各数据归零，进行新的循环判断
                Bytenum=0
                FrameState=0
        elif FrameState==2: # gyro
            
            if Bytenum<10:
                GYROData[Bytenum-2]=data
                CheckSum+=data
                Bytenum+=1
            else:
                if data == (CheckSum&0xff):
                    w = get_gyro(GYROData)
                CheckSum=0
                Bytenum=0
                FrameState=0
        elif FrameState==3: # angle
            
            if Bytenum<10:
                AngleData[Bytenum-2]=data
                CheckSum+=data
                Bytenum+=1
            else:
                if data == (CheckSum&0xff):
                    Angle = get_angle(AngleData)
#                     d = list(a+w+Angle)
#                     print("a(g):%10.3f %10.3f %10.3f w(deg/s):%10.3f %10.3f %10.3f Angle(deg):%10.3f %10.3f %10.3f"%d)
                    print('w_x:=', w[0], 'w_y:=', w[1], 'w_z:=', w[2])
                    print('angle_x:=', Angle[0], 'angle_y:=', Angle[1], 'angle_z:=', Angle[2])
                    print('a_x:=', a[0], 'a_y:=', a[1], 'a_z:=', a[2])
                    for i in range(3):
                        Omega[i].append(w[i])
                    for i in range(3):
                        Ang[i].append(Angle[i])
                    for i in range(3):
                        Acc[i].append(a[i])
                CheckSum=0
                Bytenum=0
                FrameState=0
```

```python
def get_acc(datahex):  
    axl = datahex[0]                                        
    axh = datahex[1]
    ayl = datahex[2]                                        
    ayh = datahex[3]
    azl = datahex[4]                                        
    azh = datahex[5]
    
    k_acc = 16.0
 
    acc_x = (axh << 8 | axl) / 32768.0 * k_acc
    acc_y = (ayh << 8 | ayl) / 32768.0 * k_acc
    acc_z = (azh << 8 | azl) / 32768.0 * k_acc
    if acc_x >= k_acc:
        acc_x -= 2 * k_acc
    if acc_y >= k_acc:
        acc_y -= 2 * k_acc
    if acc_z >= k_acc:
        acc_z-= 2 * k_acc
    
    return acc_x,acc_y,acc_z
```

```python
def get_gyro(datahex):                                      
    wxl = datahex[0]                                        
    wxh = datahex[1]
    wyl = datahex[2]                                        
    wyh = datahex[3]
    wzl = datahex[4]                                        
    wzh = datahex[5]
    k_gyro = 2000.0
 
    gyro_x = (wxh << 8 | wxl) / 32768.0 * k_gyro
    gyro_y = (wyh << 8 | wyl) / 32768.0 * k_gyro
    gyro_z = (wzh << 8 | wzl) / 32768.0 * k_gyro
    if gyro_x >= k_gyro:
        gyro_x -= 2 * k_gyro
    if gyro_y >= k_gyro:
        gyro_y -= 2 * k_gyro
    if gyro_z >=k_gyro:
        gyro_z-= 2 * k_gyro
    return gyro_x,gyro_y,gyro_z
 
 
def get_angle(datahex):                                 
    rxl = datahex[0]                                        
    rxh = datahex[1]
    ryl = datahex[2]                                        
    ryh = datahex[3]
    rzl = datahex[4]                                        
    rzh = datahex[5]
    k_angle = 180.0
 
    angle_x = (rxh << 8 | rxl) / 32768.0 * k_angle
    angle_y = (ryh << 8 | ryl) / 32768.0 * k_angle
    angle_z = (rzh << 8 | rzl) / 32768.0 * k_angle
    if angle_x >= k_angle:
        angle_x -= 2 * k_angle
    if angle_y >= k_angle:
        angle_y -= 2 * k_angle
    if angle_z >=k_angle:
        angle_z-= 2 * k_angle
 
    return angle_x,angle_y,angle_z
```

```python
def plotserial(Omega,Ang,Acc,fig,ax):   
    ax[0].plot(Omega[0],color = 'r',label = 'wx')
    ax[0].plot(Omega[1],color = 'g',label = 'wy')
    ax[0].plot(Omega[2],color = 'b',label = 'wz')
    ax[1].plot(Ang[0],color = 'r',label = 'angx')
    ax[1].plot(Ang[1],color = 'g',label = 'angy')
    ax[1].plot(Ang[2],color = 'b',label = 'angz')
    ax[2].plot(Acc[0],color = 'r',label = 'ax')
    ax[2].plot(Acc[1],color = 'g',label = 'ay')
    ax[2].plot(Acc[2],color = 'b',label = 'az')
    ax[0].legend(loc="lower left")
    ax[1].legend(loc="lower left")
    ax[2].legend(loc="lower left")
    ax[0].grid()
    ax[1].grid()
    ax[2].grid()
    plt.show()
```

```python
port = input('please input port No. such as com6:')
baud = int(input('please input baudrate(115200 for JY61 or 9600 for JY901):'))

ser = serial.Serial(port, baud, timeout=0.5)  # ser = serial.Serial('com6',115200, timeout=0.5)
print(ser.is_open)
Time = []
Omega = [[]for _ in range(3)]
Ang = [[]for _ in range(3)]
Acc =[[]for _ in range(3)]
now = time.time()
while(time.time()-now<=60):
#     print(time.strftime("%Y-%m-%d-%H_%M_%S", time.localtime()))
#     i+=1
#     t1 = 0.01*i
#     t.append(t1)   
    time.sleep(0.01)     # 100hz
    datahex = ser.read(33)
    DueData(datahex, Omega, Ang, Acc)
    Time.append(time.time()-now)
#     time.sleep(0.01)
```

```python
fig2,ax2 = plt.subplots(nrows=1,ncols=3,figsize = (15,6))
plotserial(Omega,Ang,Acc,fig2,ax2)
plt.show()
record = np.zeros((len(Omega[0]),10),dtype = float)
for i in range(len(Omega[0])):
    record[i,0] = Time[i]
    record[i,1] = Omega[0][i]
    record[i,2] = Omega[1][i]
    record[i,3] = Omega[2][i]
    record[i,4] = Ang[0][i]
    record[i,5] = Ang[1][i]
    record[i,6] = Ang[2][i]
    record[i,7] = Acc[0][i]
    record[i,8] = Acc[1][i]
    record[i,9] = Acc[2][i]
np.savetxt('righthand_abnormal.csv',record,delimiter = ',')
```
